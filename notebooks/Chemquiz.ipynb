```python
import random
import banco_questoes

#l√≥gica principal. vou optar por usar while true com break
print("--- CHEMQUIZ ---\
\nIsso √© um quiz com perguntas sobre qu√≠mica. \
\nDigite 'i' para iniciar ou 'e' para encerrar.\n")

lista = [] #lista vazia caso o usu√°rio escolha a op√ß√£o "aleat√≥rio"
selecting = True #selecionando a dificuldade?
active = True #jogo est√° ativo?

while active: #inicia o jogo
    entrada = input("Digite aqui: ") #resposta ao texto de introdu√ß√£o
    if entrada == 'e': #encerra o jogo
        print("At√© logo!")
        active = False
    elif entrada == 'i': # eu sei que o ideal √© usar fun√ß√µes, isso √© exerc√≠cio.
        while selecting: #selecionando a dificuldade
            dificuldade = input("Selecione o n√≠vel de dificuldade: (f)√°cil \
(m)√©dio (d)if√≠cil (a)leat√≥rio ou (e)ncerrar")
            if dificuldade == 'f':
                lista = banco_questoes.perguntas_faceis
                selecting = False
            elif dificuldade == 'm':
                lista = banco_questoes.perguntas_intermediarias
                selecting = False
            elif dificuldade == 'd':
                lista = banco_questoes.perguntas_avancadas
                selecting = False
            elif dificuldade == 'a':
                lista.extend(banco_questoes.perguntas_faceis)
                lista.extend(banco_questoes.perguntas_intermediarias)
                lista.extend(banco_questoes.perguntas_avancadas)
                selecting = False
            elif dificuldade == 'e': #boa pr√°tica de UX (usu√°rio no controle)
                print("At√© logo!")
                selecting = False
                active = False
            else: #se n√£o selecionar um argumento v√°lido, pergunta de novo
                continue
        tamanho_lista = len(lista) - 1 #argumento para o m√©todo randint()
        
        while active: #seleciona uma quest√£o aleat√≥ria e puxa pergunta/resposta
            questao = lista[random.randint(0, tamanho_lista)]
            pergunta = questao['pergunta']
            resposta = questao['resposta']
            print(f"Pergunta: {pergunta}\
            \na) {questao['a']}\nb) {questao['b']}\nc) {questao['c']}\
            \nd) {questao['d']}\n")

            entrada = input("Digite aqui: ")
            if entrada == resposta: #compara o input com a resposta correta
                print("Voc√™ acertou!")
                entrada = input("Jogar novamente?(s/n)")
                if entrada == 's' or entrada == 'i':
                    continue
                elif entrada == 'n' or entrada == 'e':
                    print("At√© logo!")
                    active = False
                else:
                    continue
            elif entrada == 'e':
                active = False
            else:
                print("Resposta incorreta.")
                entrada = input("Jogar novamente?(s/n)")
                if entrada == 's' or entrada == 'i':
                    continue
                elif entrada == 'n' or entrada == 'e':
                    print("At√© logo!")
                    active = False
                else:
                    continue
    else: #se, no in√≠cio, n√£o selecionar nada v√°lido, pergunta de novo.
        continue
```
